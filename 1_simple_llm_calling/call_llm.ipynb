```python
from dotenv import load_dotenv
from  langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
llm=ChatGoogleGenerativeAI(model="gemma-4-31b-it")

response=llm.invoke("how many moons saturn have?")
print(response.text)
```

```python
response=llm.invoke([
    ["system","You are helpful assistant that answers in one line "],
    ["human","how many moons does saturn have?"]
])
print(response.text)
```

```python
prompt="give me a single tagline to save trees"
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash",temperature=0)
response=llm.invoke(prompt)
print(response.text)
```

```python
prompt ="give me a single tagline to save trees"

llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash",temperature=0.8)
response=llm.invoke(prompt)
print(response.text)
```
